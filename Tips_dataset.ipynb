```python
import seaborn as sns
%matplotlib inline
```

```python
# Load one of the data sets that come with seaborn
tips = sns.load_dataset("tips")
tips
```

```python
tips['sex2'] = tips['sex'].map( {'Female': 0, 'Male': 1} ).astype(int)
```

```python
import pandas as pd
tips.columns
```

```python
tips['perc_bill'] = (tips.tip/tips.total_bill)*100
```

```python
grouped = tips[['tip','sex']].groupby('sex')
```

```python
grouped.boxplot(grouped,return_type='dict')
```

```python

grouped2 = tips[['perc_bill','sex']].groupby('sex')
```

```python
import matplotlib.pyplot as plt
grouped2.boxplot(grouped,return_type='dict')
```
